```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("rgistro testes.csv")
df
```

```python
df.dtypes
```

```python
b_line = df[df['alg'] == 'B']
u_line = df[df['alg'] == 'U']
i_line = df[df['alg'] == 'I']
g_line = df[df['alg'] == 'G']
a_line = df[df['alg'] == 'A']
```

```python
plt.plot(b_line['tam_vetor'], b_line['n_expl'], label='B')
plt.plot(u_line['tam_vetor'], u_line['n_expl'], label='U')
plt.plot(i_line['tam_vetor'], i_line['n_expl'], label='I')
plt.plot(g_line['tam_vetor'], g_line['n_expl'], label='G')
plt.plot(a_line['tam_vetor'], a_line['n_expl'], label='A')
plt.yscale("log")
plt.title("Nmero de ns explorados para cada tamanho de entrada \n(em escala logartmica)")
plt.xlabel("Tamanho de entrada")
plt.ylabel("Nmero de ns explorados")
plt.legend()
```

```python
plt.plot(b_line['tam_vetor'], b_line['custo'], label='B')
plt.plot(u_line['tam_vetor'], u_line['custo'], label='U')
plt.plot(i_line['tam_vetor'], i_line['custo'], label='I')
plt.plot(g_line['tam_vetor'], g_line['custo'], label='G')
plt.plot(a_line['tam_vetor'], a_line['custo'], label='A')
plt.title("Custo da soluo encontrada para cada tamanho de entrada")
plt.xlabel("Tamanho de entrada")
plt.ylabel("Custo da soluo encontrada")
plt.legend()
```

```python
plt.plot(b_line['tam_vetor'], b_line['tempo'], label='B')
plt.plot(u_line['tam_vetor'], u_line['tempo'], label='U')
plt.plot(i_line['tam_vetor'], i_line['tempo'], label='I')
plt.plot(g_line['tam_vetor'], g_line['tempo'], label='G')
plt.plot(a_line['tam_vetor'], a_line['tempo'], label='A')
plt.yscale('log')
plt.title("Tempo de execuo para cada tamanho de entrada \n(em escala logartmica)")
plt.xlabel("Tamanho de entrada")
plt.ylabel("Tempo de execuo (em segs)")
plt.legend()
```
